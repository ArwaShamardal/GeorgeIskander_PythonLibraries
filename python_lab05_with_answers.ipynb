### Write a function that will plot a circle given its center and radius using Matplotlib.

```python
def circle(cx, cy, r):
    pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
def circle(cx,cy,r):
  theta=np.linspace(0,2*np.pi,200) 
  x=r*np.cos(theta)+cx
  y=r*np.sin(theta)+cy
  plt.figure(figsize=(r+1,r+1))
  plt.plot(x,y,linewidth=1,linestyle='-')
  plt.show()

#  return x,y

circle(2,0,5)
```

### Write a function that will plot a spiral given its center, maximum radius, and number of turns using Matplotlib.

```python
def spiral(cx, cy, r, turns):
    pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
def spiral(cx, cy, r, turns):
  theta=np.linspace(0,2*np.pi*turns,200) 
  d = np.linspace(0,r,200)
  x=d*np.cos(theta)+cx
  y=d*np.sin(theta)+cy
#   plt.figure(figsize=(r+1,r+1))
  plt.plot(x,y,linewidth=1,linestyle='-')
  plt.show()

#  return x,y

spiral(2,0,5,4)
```

### Student Grades Distribution
"""
Create a bar chart showing grades distribution for 5 students.
Data: Students = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
      Grades = [85, 92, 78, 95, 88]
Requirements:
- Color bars based on grade (>=90: green, >=80: yellow, <80: red)
- Add value labels on top of each bar, hint: check bar getters (e.g. bar.get_height())
- Add a horizontal line at grade 80 (passing threshold)
"""


```python
students = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
grades = [85, 92, 78, 95, 88]
```

```python
def grades_solution():
    students = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
    grades = [85, 92, 78, 95, 88]
    
    colors = ['yellow' if g >= 80 and g < 90 else 'green' if g >= 90 else 'red' for g in grades]
    
    plt.figure(figsize=(10, 6))
    bars = plt.bar(students, grades, color=colors, linewidth=1.5)
    
    # Add value labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height}',
                ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    plt.axhline(y=80, color='blue', linestyle='--', linewidth=2, label='Passing Threshold')
    plt.title('Student Grades Distribution', fontsize=16, fontweight='bold')
    plt.xlabel('Students', fontsize=12)
    plt.ylabel('Grade', fontsize=12)
    plt.ylim(0, 105)
    plt.legend()
    plt.tight_layout()
    plt.show()

grades_solution()
```

# Subplots - Weather Dashboard
"""
Question: Create a 2x2 subplot dashboard showing different weather metrics.

Requirements:
- Top-left: Temperature line plot
- Top-right: Humidity line plot
- Bottom-left: Wind speed scatter plot with size variation
- Bottom-right: Pressure bar plot
- All subplots should have titles and proper labels
"""



```python
hours = np.arange(0, 24)
np.random.seed(42)
temperature = 20 + 5*np.sin(hours/4)
humidity = 60 + 15*np.cos(hours/3)
wind_speed = 10 + 5*np.random.rand(24)
pressure = 1013 + 3*np.sin(hours/6)
```

```python
def dashboard_solution():
    hours = np.arange(0, 24)
    np.random.seed(42)
    temperature = 20 + 5*np.sin(hours/4)
    humidity = 60 + 15*np.cos(hours/3)
    wind_speed = 10 + 5*np.random.rand(24)
    pressure = 1013 + 3*np.sin(hours/6)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Temperature
    axes[0, 0].plot(hours, temperature, color='red', linewidth=2)
    axes[0, 0].set_title('Temperature Throughout Day', fontweight='bold')
    axes[0, 0].set_xlabel('Hour')
    axes[0, 0].set_ylabel('Temperature (C)')
    axes[0, 0].grid(True, alpha=0.3)
    
    # Humidity
    axes[0, 1].plot(hours, humidity, color='darkblue', linewidth=2)
    axes[0, 1].set_title('Humidity Throughout Day', fontweight='bold')
    axes[0, 1].set_xlabel('Hour')
    axes[0, 1].set_ylabel('Humidity (%)')
    axes[0, 1].grid(True, alpha=0.3)
    
    # Wind Speed
    sizes = wind_speed * 20
    axes[1, 0].scatter(hours, wind_speed, s=sizes, alpha=0.6, c=wind_speed, 
                       cmap='viridis', edgecolors='black')
    axes[1, 0].set_title('Wind Speed Throughout Day', fontweight='bold')
    axes[1, 0].set_xlabel('Hour')
    axes[1, 0].set_ylabel('Wind Speed (km/h)')
    axes[1, 0].grid(True, alpha=0.3)
    
    # Pressure
    axes[1, 1].bar(hours, pressure, color='green', alpha=0.7, edgecolor='black')
    axes[1, 1].set_title('Atmospheric Pressure Throughout Day', fontweight='bold')
    axes[1, 1].set_xlabel('Hour')
    axes[1, 1].set_ylabel('Pressure (hPa)')
    axes[1, 1].grid(True, alpha=0.3, axis='y')
    # # show number as text on top of each bar
    # for i in range(len(hours)):
    #     axes[1, 1].text(hours[i], pressure[i] + 0.5, f'{pressure[i]%1000:.1f}', 
    #                     ha='center', va='bottom', fontsize=8)
    
    plt.suptitle('24-Hour Weather Dashboard', fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

dashboard_solution()
```
